# **Question 2**

(30 points). Write a python program to **clean the text data** you collected above and save the data in a new column in the csv file. The data cleaning steps include:

(1) Remove noise, such as special characters and punctuations.

(2) Remove numbers.

(3) Remove stopwords by using the [stopwords list](https://gist.github.com/sebleier/554280).

(4) Lowercase all texts

(5) Stemming. 

(6) Lemmatization.

```
import csv
import re
import nltk
import os
import textwrap
import pandas as pd
import numpy as np
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize
from nltk.stem.wordnet import WordNetLemmatizer
nltk.download('punkt')
nltk.download('wordnet')
import urllib.request
import requests

#temp_data = [] # Temporary list to store read data
ps = PorterStemmer()
stem_list_of_words = []

lemmatizer = WordNetLemmatizer()
lemma_list_of_words = []

#Reading list of stop words
stoplist ="/content/stopwords.txt"
with open(stoplist) as file_object:
  finalStopList =file_object.read()

# Open output file for reading
with open("reviews.csv",'r') as input_file:
    file_reader = csv.reader(input_file) # Create file reading variable
    for row in file_reader: # Loop to read each row from the file
      clean_line = re.sub(r"[^a-zA-Z\s\']","",row[1])
      clean_line = re.sub("\s\s"," ",clean_line)
     
      clean_line = re.sub(r'[0-9]+', '',clean_line) # removing numbers
      clean_line = re.sub('[A-Z]+', lambda m: m.group(0).lower(),clean_line) # making lowercase

      clean_line = [word for word in clean_line.split() if word not in finalStopList] # removing stopwords
      #print(clean_line)
      
      s2 = ' '.join([ps.stem(word) for word in clean_line])
      stem_list_of_words.append(s2)
      l2 = ' '.join([lemmatizer.lemmatize(word) for word in clean_line]) # doing lemmatization
      lemma_list_of_words.append(l2)


      f=open('f1.txt','w')
      for ele in lemma_list_of_words:
        f.write(ele+'\n')
      f.close()
 
      df = pd.read_csv("f1.txt",delimiter=',')
      df.to_csv('Demo1.csv')
      df
      
     
```
